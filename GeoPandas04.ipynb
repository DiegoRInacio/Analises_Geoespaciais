<h3>Criando mapas de agrupamento <stronger>(clusters)</stronger></h3>

Para mostrar todos os outros dados é melhor realizar dessa forma para uma melhor amostragem e processamento dos dados.

Para isso vamos ultilizar um plugin do folium para melhorar a visualização dos dados realizando um agrupamento dinâmico.



```python
fmap = folium.Map(location=[media_latitude, media_longitude])

locais = gdf_niteroi_ocorr[['latitude', 'longitude']].values.tolist()

mc = FastMarkerCluster(data=locais)

fmap.add_child(mc)

fmap
```

***
<h2>Como nós podemos melhorar a visualização desse mapa?</h2>

  - 1. Trocar o mapa base;
  - 2. Adicionar os limites do municipio de Niterói RJ
  - 3. Realizar uma mudança no estilo dos dados.

```python
fmap = folium.Map(location=[media_latitude, media_longitude])

locais = gdf_niteroi_ocorr[['latitude', 'longitude']].values.tolist()

limite = folium.features.GeoJson(gdf_limite_Nit,
                                 style_function=lambda featura: {
                                     'color': 'black',
                                     'weght': 2,
                                     'fillOpacity': 0.0
                                 })

fmap.add_child(limite)

mc = FastMarkerCluster(data=locais)

fmap.add_child(mc)

fmap
```

<h3>Como posso salvar o meu mapa para apresentar para equipe</h3>

```python
fmap = folium.Map(location=[media_latitude, media_longitude],tiles = 'cartodbpositron')

locais = gdf_niteroi_ocorr[['latitude', 'longitude']].values.tolist()

limite = folium.features.GeoJson(gdf_limite_Nit,
                                 style_function=lambda featura: {
                                     'color': 'black',
                                     'weght': 2,
                                     'fillOpacity': 0.0
                                 })

fmap.add_child(limite)

mc = FastMarkerCluster(data=locais)

fmap.add_child(mc)

fmap
```

```python
#salvando o mapa para apresentação

fmap.save('/content/drive/MyDrive/Diego Faculdade/Estudos sobre DataSciense/Resultados em html/Ocorrencias_Niteroi.html')
```
